```python
import pandas as pd
from google.colab import files
import io

uploaded = files.upload()
uploaded1 = files.upload()
```

```python
train = pd.read_csv(io.BytesIO(uploaded['train (1).csv']))
test = pd.read_csv(io.BytesIO(uploaded1['test (1).csv']))
```

```python
train.head()
```

```python
train.shape
```

```python
train.isna().sum()
```

```python
train = train.drop('Cabin' , axis = 1)
train = train.drop('PassengerId' , axis = 1)
train = train.drop('Name' , axis = 1)
train.head()
```

```python
train.dtypes
```

```python
# train['CryoSleep'] = train['CryoSleep'].astype('string')
# train['HomePlanet'] = train['HomePlanet'].astype('string')
# train['Destination'] = train['Destination'].astype('string')
# train['VIP'] = train['VIP'].astype('string')

# train.head()
```

```python
train['HomePlanet'].fillna('OtherPlanet' , inplace = True)
train['Destination'].fillna('OtherDestination' , inplace = True)
train['CryoSleep'].fillna(False , inplace = True)
train['VIP'].fillna(False , inplace = True)

lst = [ 'Age' , 'RoomService' , 'FoodCourt' , 'ShoppingMall' , 'Spa' , 'VRDeck']

for l in lst :
  train[l].fillna(train[l].mean() , inplace = True)

train.isna().sum()
```

```python
train['CryoSleep'].mode()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder =  LabelEncoder()

train['HomePlanet'] = label_encoder.fit_transform(train['HomePlanet'])
train['Destination'] = label_encoder.fit_transform(train['Destination'])
# train['Cryosleep'] = label_encoder.fit_transform(train['Cryosleep'])
# train['VIP'] = label_encoder.fit_transform(train['VIP'])
train['Transported'] = label_encoder.fit_transform(train['Transported'])

train.head()
```

```python
train['CryoSleep'] = label_encoder.fit_transform(train['CryoSleep'])
train['VIP'] = label_encoder.fit_transform(train['VIP'])
```

train.head()

```python
train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

mnmx = MinMaxScaler()
for l in lst :
  train[l] = mnmx.fit_transform(train[[l]])

train.head()
```

```python
y = train['Transported']
x = train.drop('Transported' , axis = 1)

y.head()
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(x,y)

logreg.score(x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dectree = DecisionTreeClassifier()
dectree.fit(x,y)

dectree.score(x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
randomforest = RandomForestClassifier(n_estimators = 100)
randomforest.fit(x, y)

score = randomforest.score(x, y)
print(score)
```
