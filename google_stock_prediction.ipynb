```python
from google.colab import files
import io
import pandas as pd

upload = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(upload['GOOG.csv']))
df.head()
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'].apply(lambda x : x.split()[0]))
df
```

```python
from matplotlib import pyplot as plt

columns_to_plot = ['close', 'open',]
fig = plt.figure(figsize=(10, 10))
k = 1
for i in columns_to_plot :
  plt.subplot(3 , 2 , k)
  plt.plot(df['date'] , df[i])
  plt.title(i)
  k = k+1
plt.show()
# plt.plot(df['date'] , df['high'] ,   color = 'red')
# plt.plot(df['low'] , df['date'] , color = 'blue' )
```

```python
data = df[['date' , 'open' , 'close']]
data.set_index('date' , drop = True , inplace = True)
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data['open'] = scaler.fit_transform(data[['open']])
data['close'] = scaler.fit_transform(data[['close']])

data
```

```python
train_size = round(len(data)*0.80)
train_size
```

```python
train = data[:train_size]
test = data[train_size:]
```

```python
import numpy as np
def create_seq(dataset) :
  seq = []
  labels = []
  startidx = 0

  for i in range(50 , len(dataset)) :
    seq.append(dataset.iloc[startidx : i])
    labels.append(dataset.iloc[i])
    startidx = startidx + 1
  return (np.array(seq) , np.array(labels))
```

```python
train_seq , train_label = create_seq(train)
test_seq , test_label = create_seq(test)
train_seq.shape, train_label.shape, test_seq.shape, test_label.shape
```

```python
from keras import Sequential
from keras.layers import Dense , LSTM

model = Sequential()
model.add(LSTM(50 , return_sequences=True,input_shape = (train_seq.shape[1], train_seq.shape[2])))
model.add(LSTM(50))
model.add(Dense(2))

model.compile(optimizer = 'adam'  , loss = 'mean_squared_error' , metrics = ['mean_absolute_error'])
```

```python
model.fit(train_seq , train_label , epochs = 80 , validation_data=(test_seq, test_label))
```

```python
test_predict = model.predict(test_seq)
test_inverse = scaler.inverse_transform(test_predict)
```

```python
gs_slic_data = pd.concat([data.iloc[-202:].copy(),pd.DataFrame(test_inverse,columns=['open_predicted','close_predicted'],index=data.iloc[-202:].index)], axis=1)
gs_slic_data[['open','close']] = scaler.inverse_transform(gs_slic_data[['open','close']])
```

```python
gs_slic_data[['open','open_predicted']].plot(figsize=(10,6))
plt.xticks(rotation=45)
plt.xlabel('Date',size=15)
plt.ylabel('Stock Price',size=15)
plt.title('Actual vs Predicted for open price',size=15)
plt.show()
```

```python
gs_slic_data[['close','close_predicted']].plot(figsize=(10,6))
plt.xticks(rotation=45)
plt.xlabel('Date',size=15)
plt.ylabel('Stock Price',size=15)
plt.title('Actual vs Predicted for close price',size=15)
plt.show()
```

**predicting next 10 days**

```python

gs_slic_data = gs_slic_data.append(pd.DataFrame(columns=gs_slic_data.columns,index=pd.date_range(start=gs_slic_data.index[-1], periods=11, freq='D', closed='right')))
```

```python
upcoming_prediction = pd.DataFrame(columns=['open','close'],index=gs_slic_data.index)
upcoming_prediction.index=pd.to_datetime(upcoming_prediction.index)
# upcoming_prediction.index=pd.to_datetime(upcoming_prediction.index)
```

```python
curr_seq = test_seq[-1:]

for i in range(-10,0):
  up_pred = model.predict(curr_seq)
  upcoming_prediction.iloc[i] = up_pred
  curr_seq = np.append(curr_seq[0][1:],up_pred,axis=0)
  curr_seq = curr_seq.reshape(test_seq[-1:].shape)
```

```python
upcoming_prediction[['open','close']] = scaler.inverse_transform(upcoming_prediction[['open','close']])
# upcoming_prediction
```

```python
fg,ax=plt.subplots(figsize=(10,5))
ax.plot(gs_slic_data.loc['2021-04-01':,'open'],label='Current Open Price')
ax.plot(upcoming_prediction.loc['2021-04-01':,'open'],label='Upcoming Open Price')
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
ax.set_xlabel('Date',size=15)
ax.set_ylabel('Stock Price',size=15)
ax.set_title('Upcoming Open price prediction',size=15)
ax.legend()
fg.show()
```
