```python
import pandas as pd
from google.colab import files
import io

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['train.csv']))
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop('id' , axis = 1)
```

```python
y = df[['EC1' , 'EC2','EC3' , 'EC4','EC5' , 'EC6']]
y.head()

target_1 = y['EC1']
target_2 = y['EC2']
```

```python
x = df.drop(['EC1' , 'EC2','EC3' , 'EC4','EC5' , 'EC6','SMR_VSA10','HallKierAlpha','FpDensityMorgan3','FpDensityMorgan2','FpDensityMorgan1','Kappa3','fr_COO','fr_COO2','EState_VSA2'] , axis = 1)
x.head()
```

```python
x.iloc[0]
```

```python
from joblib.compressor import XZCompressorWrapper
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = (x-x.min())/(x.max()-x.min())



```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate

dec = DecisionTreeClassifier()

cross = cross_validate(dec , X , target_1 , cv = 5)

cross['test_score']
# dec.fit(x,y)

# dec.score(x,y)

# dec.predict_proba([x.iloc[0]])
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100)

cross = cross_validate(rf , X , target_1 , cv = 5)

cross['test_score']
```

```python
from sklearn.linear_model import RidgeClassifier

rc = RidgeClassifier()

cross = cross_validate(rc , X , target_1 , cv = 5)

cross['test_score']
```

```python
cross
```

```python
from sklearn.neighbors import KNeighborsClassifier

knc = KNeighborsClassifier()

cross = cross_validate(knc , X , target_1 , cv = 5)

cross['test_score']
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators = 100 , learning_rate = 0.05)

cross = cross_validate(xgb , X , target_1 , cv = 5)

cross['test_score']
```

```python
xgb.fit(X , target_1)
first = pd.DataFrame(xgb.predict_proba(X))
first.head()
```
