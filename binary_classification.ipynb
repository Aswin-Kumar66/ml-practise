```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['train.csv']))
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
a = (df[df['Machine failure'] == 0])
b = (df[df['Machine failure'] == 1])
some_data = a.sample(2148)
```

```python
some_data
```

```python
new_train = pd.concat([some_data , b] )
new_train
```

```python
new_train.isna().sum()
```

```python
new_train = new_train.drop(['id' , 'Product ID'] ,axis = 1)
```

```python
dummy = pd.get_dummies(new_train['Type'])
dummy
```

```python
new_train = pd.concat([dummy , new_train] , axis  = 1)

new_train.isna().sum()
```

```python
new_train = new_train.drop('Type' , axis = 1)
```

```python
new_train
```

```python
new_train.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
new_train['Air temperature [K]'] = scaler.fit_transform(new_train[['Air temperature [K]']])
new_train['Process temperature [K]'] = scaler.fit_transform(new_train[['Process temperature [K]']])
new_train['Rotational speed [rpm]'] = scaler.fit_transform(new_train[['Rotational speed [rpm]']])
new_train['Torque [Nm]'] = scaler.fit_transform(new_train[['Torque [Nm]']])
new_train['Tool wear [min]'] = scaler.fit_transform(new_train[['Tool wear [min]']])
new_train
```

```python
new_train = new_train.sample(frac =1 )
print(new_train.shape)
new_train
```

```python
x = new_train.drop('Machine failure' , axis = 1)
y = new_train['Machine failure']
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
# from sklearn.ensemble import XGBoost

log = LogisticRegression()
dec = DecisionTreeClassifier()
svm = SVC()
rf = RandomForestClassifier(n_estimators= 50)
# xgb = XGBoost()

models = [log , dec , svm , rf ]

for model in models :
  model.fit(x,y)
  print(model , ':' , model.score(x,y))
```
